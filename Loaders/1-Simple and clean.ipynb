```python
from langchain_community.document_loaders import (
    TextLoader,
    PyPDFLoader,
    UnstructuredHTMLLoader,
    DirectoryLoader,
)

from pathlib import Path

# data_dir = r"C:\MYROOT\NOTEBOOKS\Ollama-RAG\data"                  # reliable
data_dir = Path(r"C:\MYROOT\NOTEBOOKS\Ollama-RAG\data")              # safest
```

```python
docs = []

docs.extend(
    DirectoryLoader(
        #"data",
        data_dir,
        glob="**/*.txt",
        loader_cls=TextLoader,
        show_progress=True,
    ).load()
)
```

```python
docs.extend(
    DirectoryLoader(
        #"data",
        data_dir,
        glob="**/*.pdf",
        loader_cls=PyPDFLoader,
        show_progress=True,
    ).load()
)

```

```python
docs.extend(
    DirectoryLoader(
        #"data",
        data_dir,
        glob="**/*.html",
        loader_cls=UnstructuredHTMLLoader,
        show_progress=True,
    ).load()
)

```

```python
print(f"Loaded {len(docs)} documents")
```
